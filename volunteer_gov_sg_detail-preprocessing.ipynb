```python
import pandas as pd
import numpy as np
```

```python
file = pd.read_csv('./volunteer_gov_sg_detail.csv',index_col=0)
```

```python
file
```

```python
grouped = file.groupby('url').agg({'Available Slots': np.median, 'Location': pd.Series.nunique})
```

```python
grouped
```

```python
singles = grouped[grouped['Location'] == 1]
```

```python
singles
```

```python
import re
```

```python
def find_postal(x):
    return re.findall('Singapore \d{6}',x)[0]
    
```

```python
locs = file[file['url'].isin(singles.index)]['Location']
```

```python
postal = file[file['url'].isin(singles.index)]['Location'].apply(find_postal)
```

```python
postal
```

```python
file['Location Postal'] = postal
```

```python
file['Location Postal'] = file['Location Postal'].fillna('Multiple Locations')
```

```python
grouped = file.groupby('url').agg({'Available Slots': np.median, 'Location Postal': pd.Series.mode})
```

```python
grouped.to_json('./export.json',orient='index')

```
